## **Airplane Satistafaction**

### Tampilkan masing masing data

```python
#data training
df_train = pd.read_csv('train.csv')
df_train.head()
```

```python
# data testing
df_testing = pd.read_csv('test.csv')
df_testing.head()
```

### Tampilkan masing masing dimensi data

```python
# data training 
print('dimensi data training', df_train.shape)
print('dimensi data testing', df_testing.shape)
```

### Check Null Value

```python
print (df_train.isnull().sum())
print('='*50)
print('='*50)
print (df_testing.isnull().sum())
```

**melihat dari komposisi diatas, untuk null value lebih baik agar di drop saja dikarenakan tidak mencapai 10% dari keseluruhan data sehingga tidak mempengaruhi keseluruhan hasil prediksi nantinya**

```python
# drop null value
new_train = df_train.dropna(axis=0)
new_test = df_testing.dropna(axis=0)
```

### Preprocessing

```python
new_train = new_train.replace({'neutral or dissatisfied': 0, 
                               'satisfied':1})
```

```python
new_train.satisfaction.value_counts()
```

```python
hasil_train = pd.get_dummies(new_train, prefix = ['Gender', 'Customer Type', 'Type of Travel', 'Class'])
hasil_train.head()
```

```python
# preprocessing dummies split and drop useless columns
x = hasil_train.drop(columns = ['satisfaction', 'Unnamed: 0', 'id'])
y = hasil_train.satisfaction
```

```python
# splitting
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=.8, random_state=42)
```

```python
X_train
```

### Random Forest

```python
rf = RandomForestClassifier(n_estimators=200, max_depth=5)
rf.fit(X_train, y_train)
```

### Model Prediction

```python
y_pred  = rf.predict(X_test)
print(classification_report(y_test, y_pred))
```

### New Data Prediction

```python
new_data_baru = new_test.replace({'neutral or dissatisfied': 0, 
                               'satisfied':1})
```

```python
new_data_baru.head()
```

```python
data_baru = new_data_baru.drop(columns = ['satisfaction', 'Unnamed: 0', 'id'])
data_baru = pd.get_dummies(data_baru, prefix = ['Gender', 'Customer Type', 'Type of Travel', 'Class'])
data_baru.head()
```

```python
# hasil prediksi data baru
hasil_prediksi = rf.predict(data_baru)
print(classification_report(hasil_prediksi, new_data_baru['satisfaction']))
```
